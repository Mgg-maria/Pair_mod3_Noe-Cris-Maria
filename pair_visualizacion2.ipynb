1. ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.

```python
import pandas as pd
```

```python
df = pd.read_csv("./world_data_full_apply_nulos.csv", index_col = 0)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Paso 1: Crear el gráfico de barras (Countplot)

plt.figure(figsize=(8, 6))  # Ajusta el tamaño del gráfico
sns.countplot(x='continent', data=df, order=df['continent'].value_counts().index, palette='Set2')  

# Mostrar el gráfico
plt.show()
```

2. ¿Cuáles son los 10 países con la mayor área forestal?

```python
df.head()
```

```python
# Ordenar los países por área forestal en orden descendente y seleccionar los 10 primeros

top_10_forest = df[['country', 'land_area']].sort_values(by='land_area', ascending=False).head(10)

# Mostrar los 10 países con mayor área forestal
print(top_10_forest)
```

```python
# Crear un gráfico de barras horizontal

plt.figure(figsize=(10, 6))
sns.barplot(x='land_area', y='country', data=top_10_forest, palette='viridis')

# Títulos y etiquetas
plt.title('Top 10 países con mayor área forestal')
plt.xlabel('Área Forestal (en km²)')  
plt.ylabel('País')

# Mostrar el gráfico
plt.show()
```
