```python
#Realiza un análisis de la mortalidad infantil (infant_mortality) y clasifica los países en tres categorías: baja, moderada y alta mortalidad infantil.

# Función para clasificar según los valores de mortalidad infantil
def clasificar_mortalidad(valor):
    if valor <= 10:
        return "Baja"
    elif valor <= 30:
        return "Moderada"
    else:
        return "Alta"

# Aplicamos la función a la columna 'infant_mortality'
df['infant_mortality_category'] = df['infant_mortality'].apply(clasificar_mortalidad)

df['infant_mortality_category']
```

¿Qué hace esta línea?
df["fertility_rate"] y df["life_expectancy"]:

Aquí estamos seleccionando dos columnas del DataFrame df:

fertility_rate: La tasa de fertilidad (número de hijos por mujer en edad fértil).

life_expectancy: La esperanza de vida (número promedio de años que se espera que viva una persona en un país).

corr():

El método .corr() en pandas calcula la correlación entre dos series de datos (en este caso, entre las columnas fertility_rate y life_expectancy).

La correlación es una medida estadística que describe la relación entre dos variables.

El valor de la correlación varía entre -1 y 1:

1: Correlación perfecta positiva (a medida que una variable aumenta, la otra también lo hace).

-1: Correlación perfecta negativa (a medida que una variable aumenta, la otra disminuye).

0: No hay correlación entre las variables (cambios en una variable no afectan a la otra).

¿Qué significa el resultado?

La correlación puede ayudarte a entender cómo se relacionan la tasa de fertilidad y la esperanza de vida. Por ejemplo:

Si la correlación es positiva, significaría que países con mayor tasa de fertilidad tienden a tener una mayor esperanza de vida (o viceversa).

Si la correlación es negativa, indicaría que en países con mayor tasa de fertilidad, la esperanza de vida tiende a ser más baja.

Si la correlación es cercana a 0, significa que no hay una relación directa clara entre ambas variables.
